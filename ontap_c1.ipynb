```python
#khai bao lóp sinh vien 
class SinhVien:
    def __init__(self, ten,lop, mssv):
        self.ten = ten
        self.lop = lop
        self.mssv = mssv
    
    #phuong thuc info()
    def info(self):
        print("thông tin sinh viên:")
        print(f"Tên: {self.ten}, MSSV: {self.mssv}, lop:{self.lop}")

class svk18a3(SinhVien):
    #khong co init
    #ghi đè phương thức info()
    def info(self):
        print("Thông tin sinh viên K18A3:")
        print(f"Tên: {self.ten}, MSSV: {self.mssv}, lop:{self.lop}")

sv = svk18a3("Nguyen Van A", "dhkl18a3", "24174600172")
sv.info()
```
